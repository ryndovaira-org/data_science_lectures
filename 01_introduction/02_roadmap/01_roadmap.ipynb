## Что НЕ включено в программу курса?

- Подробное описание внутреннего устройства алгоритмов (моделей), **только основные концепции алгоритмов**.


- Описания ВСЕХ моделей для всех типов данных, только самые знаменитые.


- Информации на темы:
    - Какая модель/метрика/библитека самая лучшая? - Все они важны для разных случаев.
    - Как сделать так, чтобы было 100% качество (~идеально)? - Нереалистично.


- Никакой "магии" AI и мифов от маркетологов/продажников. Реальность: 
    - Английский ОЧЕНЬ важен для DS.
    - AI пока что далек от захвата мира.
    - DS не "роботопсихолог" (даже если у него есть диплом психолога).
    - DS не ходит в халате и не использует пробирки.
    - DS не получает "бесконечное количество денег", только в исключительных случаях (PhD / победители мировых совервнований).
    - DS - это не простая работа, требует внимательности и дотошности (анализ данных).
    - DS - это не только модели, наоборот, модели - это малая часть работы DS.
    - DS - это серьзная работа, которая может вызывать много неприятных эмоций (нелегко получить хорошие результаты). Желательно с самого начала смириться с тем, что ошибки и неудачи будут встречаться - это абсолютно нормально.
