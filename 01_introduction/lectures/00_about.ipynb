## Как будут проходить занятия?


- 12.00-13.00 + 15м перерыв + 13.15-14.15


- Домашние задания (с ответами для самопроверки)


- Контрольные домашние задания (без ответов)


- Для успешного завершения курса необходимо выполнить 80% контрольных домашних заданий (набрать необходимое количество баллов).


- **Во время проведения лекции задавать вопросы голосом** в любой момент. 
    - На все вопросы по курсу ответы будут даны сразу же после вопроса.
    - Все вопросы **продвинутого уровня** (которые не входят в материалы лекции) будут записаны преподавателем. **Ответы** будут даны **в общем чате в ближайшие дни**.


- **Во время проведения лекции НЕ использовать чат**. Можно использовать чат только для:
    - Экстренных случаев (не слышно/не видно).
    - Приветствия / Прощания (если хочется).
    - Ответа вида `+`/`-` на вопрос преподавателя.
    - Если *совершенно нет возможности* включить микрофон (сломался/отсутствует).


- **Вопросы по курсу** желательно задавать **в общем чате:**
    - по лекциям
    - по ДЗ
    - по настройке окружения
    - по расписанию лекций (перенос/праздничные выходные/время проведения)
    - по любым другим вопросам, которые могут быть полезны другим студентам


- **Лично преподавателю** можно задавать любые **другие вопросы**:
    - помощь с личными проектами
    - рекомендации по поиску работы
    - предложения по улучшению курса
    - обнаружение ошибок/неточностей в курсе
    - пожелания по материалам лекций
    - просто поболтать


- Рекомендации для чатов:
    - **обязательно: спокойный стиль** (без "аааа ничего не получается", "это конец!" - это не продуктивно)
    - обязательно: если нужна срочная помощь преподавателя, то написать в личные сообщения призыв посмотреть в общий чат.
    - **желательно: описать проблему минимальным количеством сообщений** (может быть несколько абзацев в одном сообщении)
    - желательно: приложить картинку проблемы (если есть)
    - желетельно: вставлять ссылки правильным способом (с названием)
