### Группировка и агрегирование

```python
# среднее значение в каждом столбце по hotel
df.groupby(["hotel"]).mean()
```

```python
# количество значений для каждой колонки относительно arrival_date_month
df.groupby("arrival_date_month").count()
```

```python
# подсчитать, сколько всего City и Resort отелей
df.groupby(['hotel'])['is_canceled'].count()
```

```python
# подсчитать, сколько было отмен (is_canceled) для City и Resort отелей
df.groupby(['hotel', "is_canceled"])['lead_time'].count()
```

```python
# подсчитать, сколько было отмен (is_canceled) у повторных поситителей и новых (is_repeated_guest) 
df.groupby(['is_repeated_guest', "is_canceled"])['lead_time'].count().unstack()
```

```python
# подсчитать, сколько было отмен (is_canceled) у повторных поситителей и новых (is_repeated_guest)
# для City и Resort отелей 
df.groupby(['hotel', "is_repeated_guest", "is_canceled"])["is_canceled"].aggregate('count').unstack()
```
