```python
import pandas as pd

df = pd.read_csv('../../data/datetime_speed.csv',
                 index_col='datetime', 
                 parse_dates=True)

df
```

```python
df.drop(columns=[df.columns[0]], inplace=True)

df
```

```python
df.sort_index(inplace=True)

df
```

```python
import matplotlib.pylab as plt
import matplotlib.dates as mdates 

plt.figure(figsize=(15, 5))

ax = df.Download.plot.line(color='green', 
                           marker='o', 
                           linestyle='dashed')

ticks = [datetime.strftime('%X\n%d.%m.%y') for datetime in df.index]
plt.xticks(df.index, ticks, rotation = 45)

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.stem(df.index, df.Download, 
         linefmt=":", 
         markerfmt="*")

ticks = [datetime.strftime('%X\n%d.%m.%y') for datetime in df.index]
plt.xticks(df.index, ticks, rotation = 45)

plt.show()
```
