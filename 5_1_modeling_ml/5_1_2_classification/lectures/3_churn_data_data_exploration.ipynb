# Исследование данных (Data exploration)

## Импорт пакетов

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

## Описание

`ChurnData.csv` (Отток клиентов):

[Dataset source](https://www.kaggle.com/gangliu/customerchurnrate)

Набор данных включает информацию о:
- Клиенты, ушедшие в течение последнего месяца - столбец называется Churn
- Услуги, на которые подписан каждый клиент - телефон, несколько линий, Интернет, онлайн-безопасность, онлайн-резервное копирование, защита устройств, техническая поддержка, а также потоковая передача ТВ и фильмов.
- Информация об учетной записи клиента - как долго он был клиентом, договор, способ оплаты, безбумажный счет, ежемесячные платежи и общие расходы
- Демографические данные о клиентах - пол, возраст, есть ли у них партнеры и иждивенцы.

### Загрузка данных

### Исследование данных (Data Exploration)

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(figsize=(15, 15))
plt.tight_layout()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(15, 15))

ax = sns.heatmap(
    correlation_matrix,
    vmax=1,
    square=True,
    annot=True,
    fmt='.2f',
    cmap='GnBu',
    cbar_kws={"shrink": .5},
    robust=True
)

plt.title('Correlation Matrix of features', fontsize=20)
```
