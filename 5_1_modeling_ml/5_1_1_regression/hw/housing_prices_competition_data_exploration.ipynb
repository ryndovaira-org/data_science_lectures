````python
cols_with_missing = [col for col in home_data.columns
                     if home_data[col].isnull().any()]
cols_with_missing
````

````python
#! conda install missingno -y

import missingno as msno
msno.matrix(home_data[cols_with_missing])
````

````python
msno.bar(home_data[cols_with_missing])
````

````python
for col_name in cols_with_missing:
    print(f'{col_name}:\n{home_data[col_name].unique()}\n')
    col_isna = home_data[col_name].isna().sum()
    col_count = home_data[col_name].count()
    print(f'isna: {col_isna}\n')
    print(f'count: {col_count}\n')
    print(f'%: {col_isna/(col_isna + col_count)*100}\n\n----------------------------------------\n')
````

### Обработка колонки `Alley`

```
Alley: Type of alley access to property

       Grvl	Gravel
       Pave	Paved
       NA 	No alley access
```

Можно заменить `NA` на `"Empty"`, а затем использовать `One-hot encoder`. 

````python
home_data['Alley'].fillna('Empty', inplace=True)
home_data['Alley']
home_data['Alley'].unique()
````

````python
alley_dummies = pd.get_dummies(home_data[['Alley']])
alley_dummies
````

````python
home_data_num = home_data.select_dtypes(include=[np.number])
home_data_num.info()
````

````python
home_data.hist(figsize = (30, 30))
plt.show()
````

````python
correlation_matrix = home_data.corr()

plt.figure(figsize=(40, 40))

ax = sns.heatmap(
    correlation_matrix,
    vmax=1,
    square=True,
    annot=True,
    fmt='.2f',
    cmap='BuPu',
    cbar_kws={"shrink": .5},
    annot_kws={"size": 15},
    robust=True
)

plt.title('Correlation Matrix of features', fontsize=20)
````
