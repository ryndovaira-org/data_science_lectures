## Подготовка данных

```python
# добавить новый столбец (=all_days), который является суммой stays_in_week_nights (=weekdays) и stays_in_weekend_nights (=weekends)
df['stays_nights'] = df.stays_in_week_nights + df.stays_in_weekend_nights
df
```

```python
meal_groupby = df.groupby('meal').sum()

meal_canceled = meal_groupby.is_canceled
meal_groupby
```

```python
month_sum = df.groupby(by=['arrival_date_month']).sum()
month_sum
```

## Графики (line plot)

```python
meal_groupby[['adults', 'children']].plot.line()
```

```python
meal_groupby[['adults', 'children']].plot(kind='line', 
                                          subplots=True, 
                                          layout=(2, 1), 
                                          figsize=(10, 3), 
                                          grid=True)
```

## Столбчатые диаграммы — вертикальные (vertical bar plot)

```python
month_sum.stays_nights.plot(kind='bar', 
                            grid=True, 
                            color='g')
```

## Столбчатые диаграммы — горизонтальные (horizontal bar plot)

```python
month_sum.stays_nights.plot(kind='barh', 
                            figsize=(3, 3), 
                            color='m')
```
