```python
correlation_matrix = df.corr()

plt.figure(figsize=(20, 20))

ax = sns.heatmap(
    correlation_matrix,
    vmax=1,
    square=True,
    annot=True,
    fmt=".2f",
    cmap="GnBu",
    cbar_kws={"shrink": 0.5},
    robust=True,
)

plt.title("Correlation Matrix of features", fontsize=8)
```
