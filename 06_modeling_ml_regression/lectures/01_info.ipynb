# Регрессия (Regression)

---

**Тип**:

Обучение с учителем (supervised learning).

---

**Примеры использования**:

- Предсказание цены дома по его характеристикам.
- Предсказание количества покупок товара в следующем месяце.
- Предсказание уровня воды в водохранилище.
- Предсказание температуры воздуха.

---

**Дано**:

Признаки (features).

Примеры:
- возраст
- площадь
- температура тела
- средний доход
- объем
- цвет
- число покупок

---

**Результат**:

Число или числовой вектор.

Примеры:
- Стоимость дома.
- Количество выделяемого углекислого газа.
- Доход клиента.
- Рост человека.

---

**Модели:**

- Линейная регрессия (Linear regression)
    - Простая линейная регрессия (Simple Linear Regression)
    - Множественная линейная регрессия (Multiple linear regression)
- Нелинейная регрессия (Non-linear regression) / Полиномиальная регрессия (Polynomial regression)
    - Квадратическая (Quadratic)
    - Кубическая (Cubic)
    - Экспоненциальная (Exponential)
    - Логарифмическая (Logarithmic)
    - Логистическая (Logistic)
    - ...
- XGBoost Regressor
- Random Forest Regressor
    
---

Геометрически определяет прямую (любой формы), наиболее близко проходящую ко всем точкам.
